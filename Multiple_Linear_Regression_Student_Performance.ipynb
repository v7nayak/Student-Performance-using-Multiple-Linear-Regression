```python
import seaborn as sns
import numpy as np
import sklearn as skl
import pandas as pd
```

```python
data = pd.read_csv("Student_Performance.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
replacement_dict = {
    'Yes': 1,
    'No': 0
}
# Replace values in the 'Category' column using the dictionary
data['Extracurricular Activities'] = data['Extracurricular Activities'].replace(replacement_dict)
```

```python
data['Extracurricular Activities'] = data['Extracurricular Activities'].replace({"Yes":1,"No":0})
```

```python
data.head()
```

```python
x = data.iloc[:,:5]
y = data.iloc[:,5]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("-------------------------------------------------------------------------")
print(f"The R2 score is: ------>>  {r2_score(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Mean Absolute Error is: ------>> \n{mean_absolute_error(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Mean Squared Error is: ---->> {mean_squared_error(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Root Mean Squared Error is: ---->> {np.sqrt(mean_squared_error(y_test,y_pred))}")
```
